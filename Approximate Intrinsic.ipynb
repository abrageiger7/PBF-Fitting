```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
intrinsic_pulse = np.load('Approximate Intrinsic Pulse.npy')
```

```python
time = np.arange(0,2048,1) * (0.0021499/2048) * 1e3 #milliseconds
```

```python
plt.xlabel('Pulse Phase')
plt.ylabel('Intensity')
plt.title('Approximate Intrinsic Pulse from Averaged 1700 MHz Pulses')
plt.plot(time, intrinsic_pulse)
```
