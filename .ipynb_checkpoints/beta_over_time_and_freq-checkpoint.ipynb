```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
mjd_list, beta_list, freq_list, gauss_width_list, pbf_width_list, low_chi_list, tau_list = np.load('betadatayay.npy')
```

```python
plt.plot(mjd_list, beta_list, '.')
```

```python
print(mjd_list)
print(beta_list)
```

```python
betas_mjd57649 = np.concatenate((beta_list[:3], beta_list[15:18],  beta_list[30:33], beta_list[45:48]))
betas_mjd56627 = np.concatenate((beta_list[3:6], beta_list[18:21], beta_list[33:36], beta_list[48:51]))
betas_mjd57629 = np.concatenate((beta_list[6:9], beta_list[21:24], beta_list[36:39], beta_list[51:54]))
betas_mjd56255 = np.concatenate((beta_list[9:12], beta_list[24:27], beta_list[39:42], beta_list[54:57]))
betas_mjd57251 = np.concatenate((beta_list[12:15], beta_list[27:30], beta_list[42:45], beta_list[57:]))

plt.title('Betas for MJD 56255')
plt.hist(betas_mjd56255)
plt.show()
plt.title('Betas for MJD 56627')
plt.hist(betas_mjd56627)
plt.show()
plt.title('Betas for MJD 57251')
plt.hist(betas_mjd57251)
plt.show()
plt.title('Betas for MJD 57629')
plt.hist(betas_mjd57629)
plt.show()
plt.title('Betas for MJD 57649')
plt.hist(betas_mjd57649)
plt.show()
```

```python
plt.plot(freq_list, beta_list, '.')
```

```python
plt.plot(freq_list, gauss_width_list * (0.0021499/2048) * 1e6, '.') #converting to micrseconds
plt.ylabel('Gauss Width Microseconds')
```

```python
plt.plot(freq_list, pbf_width_list, '.')
plt.xscale('log')
plt.yscale('log')
```

```python
plt.plot(freq_list, low_chi_list, '.')
```

```python
plt.plot(mjd_list, low_chi_list, '.')
```
